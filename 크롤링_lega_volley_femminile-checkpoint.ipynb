```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import pandas as pd
import time

options = Options()
options.add_experimental_option("detach", True)
options.add_argument("user-agent=Mozilla/5.0")
driver = webdriver.Chrome(options=options)

# Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ ÏõêÎ≥∏ Ìó§Îçî(HTML Íµ¨Ï°∞ Í∏∞Ï§Ä)
ORIGINAL_HEADERS = {
    "Best Scorers": ["Rank","Name","Team","Points","Attack","Block","Serve"],
    "Best Attackers": ["Rank","Name","Team","Points","Errors","Attempts","Avg Match","Success %","Total"],
    "Best Blockers": ["Rank","Name","Team","Blocks","Errors","Rebounds","Avg Match","Efficiency %","Total"],
    "Best Servers": ["Rank","Name","Team","Points","Errors","Attempts","Avg Match","Success %","Total"],
    "Best Receivers": ["Rank","Name","Team","Successful","Errors","Attempts","Avg Match","Success %","Total"]
}

# prefix Î∂ôÏó¨ÏÑú header Î≥ÄÌôòÌïòÎäî Ìï®Ïàò
def convert_headers(category, headers):
    prefix = category.lower().replace("best ", "").replace(" ", "_")
    
    new_headers = []
    for h in headers:
        if h in ["Rank", "Name", "Team"]:
            new_headers.append(h)  # Í≥µÌÜµ Ìï≠Î™©ÏùÄ Í∑∏ÎåÄÎ°ú
        else:
            key = h.lower()
            key = key.replace(" ", "_").replace("%", "pct")
            new_headers.append(f"{prefix}_{key}")
    return new_headers

# Í≥µÌÜµ ÌÅ¨Î°§ÎßÅ Ìï®Ïàò
def crawl_stats_page(url, category):
    print(f"\n=== ÌÅ¨Î°§ÎßÅ: {category} ===")

    driver.get(url)
    time.sleep(3)
    # input("Cloudflare Ïù∏Ï¶ù ÌõÑ Enter >>> ")

    # prefix Ï†ÅÏö©Îêú Ìó§Îçî
    headers = convert_headers(category, ORIGINAL_HEADERS[category])
    print("ÏÇ¨Ïö©ÎêòÎäî Ìó§Îçî:", headers)

    # table body
    tbody = driver.find_element(
        By.XPATH,
        '//*[@id="main-content"]/section[2]/div/div/div/div/div/table/tbody'
    )
    rows = tbody.find_elements(By.TAG_NAME, "tr")

    results = []

    for r in rows:
        cols = r.find_elements(By.TAG_NAME, "td")
        if len(cols) < len(headers):
            continue

        row_data = []

        for idx, td in enumerate(cols):
            if idx == 1:
                # Player nameÏùÄ <a> ÎÇ¥Î∂Ä
                try:
                    value = td.find_element(By.TAG_NAME, "a").text.strip()
                except:
                    value = td.text.strip()
            else:
                value = td.text.strip()

            row_data.append(value)

        row_dict = dict(zip(headers, row_data))
        row_dict["Category"] = category  # Íµ¨Î∂ÑÏö©

        results.append(row_dict)

    return pd.DataFrame(results)

# Ï†ÑÏ≤¥ Ïπ¥ÌÖåÍ≥†Î¶¨ URL
URLS = {
    "Best Scorers": "https://en.volleyballworld.com/volleyball/competitions/legavolley-femminile-a1/statistics/best-scorers/",
    "Best Attackers": "https://en.volleyballworld.com/volleyball/competitions/legavolley-femminile-a1/statistics/best-attackers/",
    "Best Blockers": "https://en.volleyballworld.com/volleyball/competitions/legavolley-femminile-a1/statistics/best-blockers/",
    "Best Servers": "https://en.volleyballworld.com/volleyball/competitions/legavolley-femminile-a1/statistics/best-servers/",
    "Best Receivers": "https://en.volleyballworld.com/volleyball/competitions/legavolley-femminile-a1/statistics/best-receivers/",
}

# Ïã§Ìñâ
all_frames = []

for cat, url in URLS.items():
    df = crawl_stats_page(url, cat)
    all_frames.append(df)

df_all = pd.concat(all_frames, ignore_index=True)
df_all.to_csv("legavolley_stats_prefixed.csv", index=False)

print("\nlegavolley_stats_prefixed.csv Ï†ÄÏû• ÏôÑÎ£å")
```
